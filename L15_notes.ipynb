```python
from sklearn.datasets import make_blobs,  load_digits
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import adjusted_rand_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import matplotlib.pyplot as plt
```

### Load data

```python
centers = [[1, 1], [-1, -1], [1, -1]]

X, labels = make_blobs(n_samples=750, centers=centers, cluster_std=0.4, random_state=42)

X
```

### Plot the blobs

```python
plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis', s=15, alpha=0.7)
plt.title("Generated blobs")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)

kmeans.fit(X)
```

```python
# look at labels from kmeans and compute ARI vs true labels
from sklearn.metrics import adjusted_rand_score
print("KMeans ARI:", adjusted_rand_score(labels, kmeans.labels_))

kmeans.labels_
```

```python
# --- Digits: load, LDA, plot, cluster/evaluate ---
from sklearn.datasets import load_digits
from sklearn.cluster import KMeans

# load
digits = load_digits()
X_digits = digits.data
y = digits.target

# dimension reduction with LDA
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X_digits, y)

# plot true labels
plt.figure(figsize=(6,6))
plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y, cmap='tab10', s=15, alpha=0.7)
plt.title("Digits LDA projection (true labels)")
plt.show()

# cluster digits with KMeans and evaluate
k_digits = KMeans(n_clusters=10, random_state=42).fit(X_digits)
print("KMeans on digits ARI:", adjusted_rand_score(y, k_digits.labels_))

plt.figure(figsize=(6,6))
plt.scatter(X_lda[:,0], X_lda[:,1], c=k_digits.labels_, cmap='tab20', s=15, alpha=0.7)
plt.title("Digits LDA projection (KMeans labels)")
plt.show()
```

### Run loop to test different values of n_clusters for K means

```python
# loop over different cluster counts with KMeans (fix capitalization and use correct labels)
num_cluster_list = [6, 8, 10, 12]
for num in num_cluster_list:
    kmeans = KMeans(n_clusters=num, random_state=42)
    kmeans.fit(X)
    score = adjusted_rand_score(labels, kmeans.labels_)
    print(num, score)
```

```python
# Activity 2: DBSCAN experiments
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import adjusted_rand_score

params = [(0.3, 10), (0.2, 5), (0.4, 5)]
plt.figure(figsize=(15,4))
for i, (eps, mins) in enumerate(params, 1):
    db = DBSCAN(eps=eps, min_samples=mins).fit(X)
    labels_db = db.labels_
    n_clusters = len(set(labels_db)) - (1 if -1 in labels_db else 0)
    plt.subplot(1, len(params), i)
    plt.scatter(X[:, 0], X[:, 1], c=labels_db, cmap='tab20', s=15, alpha=0.7)
    plt.title(f'eps={eps}, min={mins}\nclusters={n_clusters}')
plt.show()

# ARI vs true labels (noise labeled -1 included)
for eps, mins in params:
    db = DBSCAN(eps=eps, min_samples=mins).fit(X)
    print(f"DBSCAN eps={eps}, min_samples={mins} ARI:", adjusted_rand_score(labels, db.labels_))
```

```python
# repeat loop with AgglomerativeClustering
from sklearn.cluster import AgglomerativeClustering

num_cluster_list = [6, 8, 10, 12]
for num in num_cluster_list:
    agg = AgglomerativeClustering(n_clusters=num)
    agg.fit(X)
    print("Agglomerative", num, adjusted_rand_score(labels, agg.labels_))
```

```python
# repeat loop with DBSCAN variations
for eps, mins in [(0.3,10),(0.2,5),(0.4,5)]:
    db = DBSCAN(eps=eps, min_samples=mins).fit(X)
    print("DBSCAN", eps, mins, adjusted_rand_score(labels, db.labels_))
```
